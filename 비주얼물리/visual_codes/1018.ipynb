```python
from scipy import optimize
import numpy as np
import matplotlib.pyplot as plt

def f(x) :
    return x**2 + 10*np.sin(x)

x = np.arange(-10, 10, 0.1)
y = f(x)
#print(x)
#print(y)

plt.plot(x,f(x))
plt.show
```

```python
result = optimize.minimize(f,x0=0)
result
```

```python
result = optimize.minimize(f,x0=3, method="L-BFGS-B")
result
```

```python
optimize.basinhopping(f, 7)
```

```python
res = optimize.minimize(f, x0=1, bounds=((0,10), ))
res.x
```

```python
def sixhump(x) :
    return ((4-2.1*x[0]**2+x[0]**4/3.0)*x[0]**2+x[0]*x[1]+(-4+4*x[1]**2)*x[1]**2)

x= np.linspace(-2,2,300)
y= np.linspace(-1,1,300)

xg, yg = np.meshgrid(x ,y)

plt.figure()
plt.imshow(sixhump([xg,yg]),extent=[-2,2,-1,1])
plt.colorbar()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
surf=ax.plot_surface(xg,yg,sixhump([xg,yg]),rstride=1, cstride=1, linewidth=0, antialiased=False, cmap='viridis')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x,t)')
ax.set_title('Six hump Camelback funtion')

plt.show()
```

```python
##find the minima?
from scipy import optimize
res = optimize.minimize(surf, x0=1, bounds=((-2,2), ))
res.x,y
```
