# zigzag

```python
#   b      d       f      h
#a     c       e      g

import pybinding as pb
import numpy as np
import matplotlib.pyplot as plt
from numpy import linalg as LA
a = 1      #distance
V = 0      #site energy
t = 1      #hopping

R1 = [0,1]
R2 = (1/2,np.sqrt(3)/2)
R3 = (1/2 , -np.sqrt(3)/2)

energyband = []

k = np.linspace(-np.pi/(2*a), np.pi/(2*a), 100) #wave vector k


def f(R,x) :
    return np.exp(1j*np.dot([0,x],[R[0],R[1]]))

for i in range(len(k)) : 
    HAa=V
    HAb=t*(f(R2,k[i]))+t*(f(R3,k[i]))
    HAc=0
    HAd=0
    HAe=0
    HAf=0
    HAg=0
    HAh=0
    
    HBa=np.conjugate(HAb)
    HBb=V
    HBc=t*(f(R1,k[i]))
    HBd=0
    HBe=0
    HBf=0
    HBg=0
    HBh=0
    
    HCa=0
    HCb=np.conjugate(HBc)
    HCc=V
    HCd=t*(f(R2,k[i]))+t*(f(R3,k[i]))
    HCe=0
    HCf=0
    HCg=0
    HCh=0
    
    HDa=0
    HDb=0
    HDc=np.conjugate(HCd)
    HDd=V
    HDe=t*(f(R1,k[i]))
    HDf=0
    HDg=0
    HDh=0
    
    HEa=0
    HEb=0
    HEc=0
    HEd=np.conjugate(HDe)
    HEe=V
    HEf=t*(f(R2,k[i]))+t*(f(R3,k[i]))
    HEg=0
    HEh=0
   
    HFa=0
    HFb=0
    HFc=0
    HFd=0
    HFe=np.conjugate(HEf)
    HFf=V
    HFg=t*(f(R1,k[i]))
    HFh=0
        
    HGa=0
    HGb=0
    HGc=0
    HGd=0
    HGe=0
    HGf=np.conjugate(HFg)
    HGg=V
    HGh=t*(f(R2,k[i]))+t*(f(R3,k[i]))

    HHa=0
    HHb=0
    HHc=0
    HHd=0
    HHe=0
    HHf=0
    HHg=np.conjugate(HGh)
    HHh=V

    H = np.array([[HAa,HAb,HAc,HAd,HAe,HAf,HAg,HAh],                  
                  [HBa,HBb,HBc,HBd,HBe,HBf,HBg,HBh],
                  [HCa,HCb,HCc,HCd,HCe,HCf,HCg,HCh],
                  [HDa,HDb,HDc,HDd,HDe,HDf,HDg,HDh],
                  [HEa,HEb,HEc,HEd,HEe,HEf,HEg,HEh],
                  [HFa,HFb,HFc,HFd,HFe,HFf,HFg,HFh],
                  [HGa,HGb,HGc,HGd,HGe,HGf,HGg,HGh],
                  [HHa,HHb,HHc,HHd,HHe,HHf,HHg,HHh]])
    w, v = LA.eig(H)
    w.sort()
    energyband.append(w)

plt.figure()
plt.plot(k,energyband, label='energy band') 
```

```python
#armchair ... another unitcell
#cell unit
#   d   
#   c      
#b         
#a      

import pybinding as pb
import numpy as np
import matplotlib.pyplot as plt
from numpy import linalg as LA
a = 1      #distance
V = 0      #site energy
t = -3      #hopping

R0 = [0,-1]
R1 = [0,1]
R2 = (np.sqrt(3)/2,-1/2)
R3 = (np.sqrt(3)/2,1/2)
R4 = (-np.sqrt(3)/2,1/2)


energyband = []

k = np.linspace(-np.pi/(a), np.pi/(a), 100) #wave vector k


def f(R,x) :
    return np.exp(1j*np.dot([0,x],[R[0],R[1]]))

for i in range(len(k)) : 
    HAa=V
    HAb=t*(f(R1,k[i]))+t*(f(R2,k[i]))
    HAc=0
    HAd=0

    
    HBa = np.conjugate(HAb)
    HBb = V
    HBc = t*(f(R3,k[i]))
    HBd = 0

    
    HCa = 0
    HCb = np.conjugate(HBc)
    HCc = V
    HCd = t*(f(R1,k[i]))+t*(f(R2,k[i]))

    HDa = 0
    HDb = 0
    HDc = np.conjugate(HCd)
    HDd = V

    H = np.array([[HAa,HAb,HAc,HAd],                  
                  [HBa,HBb,HBc,HBd],
                  [HCa,HCb,HCc,HCd],
                  [HDa,HDb,HDc,HDd],])
    w, v = LA.eig(H)
    w.sort()
    energyband.append(w)

plt.figure()
plt.plot(k,energyband, label='energy band') 
```
