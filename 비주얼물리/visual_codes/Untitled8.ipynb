Matplotlib object-oriented API

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'r')

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes.plot(x,y,'r')

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')

axes2.plot(x,y,'g')
axes2.set_xlabel('x')
axes2.set_ylabel('y')
axes2.set_title('title2')
```

```python
fig.ax = plt.subplots(nrows=1, ncols=2)

for axs in ax :
    axs.plot(x,y,'r')
    axs.set_xlabel('x')
    axs.set_ylabel('y')
    axs.set_title('title')
    
fig.tight_layout()

os.chdir("/home/comphy/desktop")
fig.savefig("filename.png")
```

```python
n = np.array([0, 1, 2, 3, 4, 5])
xx = np.linspace(-0.75, 1., 100)
```

```python
fig, axes = plt.subplots(1,4, figsize=(12, 3))

axes[0].scatter(xx, xx*0.25*np.random.random(len(xx)))
axes[0].set_title("scatter")

axes[1].step(n,n**2, lw=2)
axes[1].set_title("step")

axes[2].bar(n, n**2, align="center", width=0.5, alpha=0.5)
axes[2].set_title("bar")

axes[3].fill_between(x, x**3, color="green", alpha=0.5)
axes[3].set_title("fill betwenn")
```
