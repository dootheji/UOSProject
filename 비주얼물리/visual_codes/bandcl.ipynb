```python
#two coupling 
#A A'

# EA + 2*cos(kx*a)
import numpy as np
from numpy import linalg as LA
import matplotlib.pyplot as plt

t  = -2  #eV
tv  = 1  #Ao

d=1
dv=2

EA  = 0
EAp = 0 #prime

R1 = d
R2 = -d

xk = np.linspace(-np.pi/d,np.pi/d,1000)
yk = np.linspace(-np.pi/dv,np.pi/dv,1000)
Ev = []

for i in range(1000):
    kx = xk[i]
    ky = yk[i]
    
    sf = 2*np.cos(kx*d)
    sfv= np.exp(1j*ky*dv)
    
    H  = np.array([[EA+t*sf, tv*sfv],[np.conjugate(tv*sfv), EAp+t*sf]])
    Eval,Evec  = LA.eig(H) # value, vecter
    Ev.append(Eval)
plt.plot(xk,Ev,'.r')

plt.show()
```

```python
#4 coupling 
#A A' B B'

# EA + 2*cos(kx*a)
import numpy as np
from numpy import linalg as LA
import matplotlib.pyplot as plt

t  = -2  #eV
tv  = 1  #Ao

d=1
dv=2

EA  = -1
EAp = -1 #prime
EB  = 1
EBp = 1

R1 = d
R2 = -d

xk = np.linspace(-np.pi/d,np.pi/d,1000)
yk = np.linspace(-np.pi/dv,np.pi/dv,1000)
Ev = []


for i in range(1000):
    kx = xk[i]
    ky = yk[i]
    
    sf = 2*np.cos(kx*d)
    sfv= np.exp(1j*ky*dv)
    
    H  = np.array([[EA+t*sf, t*sf, tv*sfv, 0]
                   ,[np.conjugate(t*sf), EB+t*sf, 0, tv*sfv]
                   ,[np.conjugate(tv*sfv), 0, EAp+t*sf, t*sf]
                   ,[0, np.conjugate(tv*sfv), np.conjugate(t*sf), EBp+t*sf]])
    Eval,Evec  = LA.eig(H) # value, vecter
    Ev.append(Eval)
plt.plot(xk,Ev,'.r')

plt.show()
```
